```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

# Filter sizes

```python
filters = pd.read_csv('filters.csv', sep = ';', decimal = ',')
```

```python
plt.figure(dpi=150)
ax = sns.boxplot(data = filters[filters.columns[0:4]], color = 'orange')
ax.set_xticklabels(['(A)','(B)','(C)','(D)'])
plt.title('MSE', fontsize = 20)
plt.ylim(0,0.03)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.show()
```

```python
plt.figure(dpi=150)
ax = sns.boxplot(data = filters[filters.columns[4:8]], color = 'orange')
ax.set_xticklabels(['(A)','(B)','(C)','(D)'])
plt.title('MAE', fontsize = 20)
plt.ylim(0.05,0.12)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.show()
```

```python
plt.figure(dpi=150)
ax = sns.boxplot(data = filters[filters.columns[8:12]], color = 'orange')
ax.set_xticklabels(['(A)','(B)','(C)','(D)'])
plt.title('$R^2$', fontsize = 20)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.show()
```
