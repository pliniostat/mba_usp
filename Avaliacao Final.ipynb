### Questão 3 (3 pontos)
Alguns países podem aparecer mais de uma vez no DataFrame, pois podem estar fragmentados em estados ou províncias. Sabendo disso, quais os 5 países que apresentaram o maior número de casos considerando a soma de todos os seus estados e províncias na data mais recente de registro (coluna `3/9/22`)?

```python
print('Os cinco países procurados (em ordem crescente) são: \n')
covid.groupby('Country/Region')[covid.columns[-1]].agg('sum').sort_values(ascending = False).head()
```

### Questão 4 (3 pontos)
Faça o gráfico da evolução do número acumulado de contaminados no Brasil (Brazil) e na Itália (Italy) ao longo do tempo. 

```python
# armazenando as datas
datas = pd.to_datetime(covid.columns[3:covid.shape[1]])
# valores acumulados do Brasil
br = covid.drop(['Country/Region', 'Lat', 'Long'], axis = 1).loc[covid['Country/Region'] == 'Brazil'].values
br = br.reshape(len(datas),)
# valores acumulados da Itália
it = covid.drop(['Country/Region', 'Lat', 'Long'], axis = 1).loc[covid['Country/Region'] == 'Italy'].values
it = it.reshape(len(datas),)
```

```python
import matplotlib.pyplot as plt

plt.plot(datas, br, label = 'Brasil')
plt.plot(datas, it, label = 'Itália')
plt.legend(fontsize = 12)
plt.xticks(rotation = 45)
plt.xlabel('Data (ano-mês)', fontsize = 12)
plt.ylabel('Número de casos (x10 Milhões)', fontsize = 12)
plt.show()
```
